```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
# Choose a pre-trained English -> French model
model_name = 'Helsinki-NLP/opus-mt-en-fr'
```

More details on the model - https://huggingface.co/Helsinki-NLP/opus-mt-en-fr

```python

# Load tokenizer and model
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
# Function to translate a sentence
def translate_en_to_fr(sentence):
    # Tokenize the input sentence
    inputs = tokenizer(sentence, return_tensors="pt", padding=True)

    # Generate translation using the model
    translated = model.generate(**inputs, max_length=50)

    # Decode the translated tokens
    fr_sentence = tokenizer.decode(translated[0], skip_special_tokens=True)
    return fr_sentence
```

```python
# Example sentences
sentences = [
    "Hello, how are you?",
    "I love learning new things.",
    "Transformers are powerful for NLP tasks."
]

# Translate
for s in sentences:
    print(f"EN: {s}")
    print(f"FR: {translate_en_to_fr(s)}\n")
```
